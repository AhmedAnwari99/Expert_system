**Introduction**


This is a system devised for a dental camp that was hosted in a village where there was no dental set up available. The camp aimed to diagnose and treat oral health problems in the population of the village. Since there was no established dental set up we anticipated a large number of patients to show up. We were a group of 5 dentists, and three helpers, one of whom was in charge of noting down the symptoms/ complaints of the patients. The symptoms entered into the expert system defined which of the 5 dentists the patient was going to be referred to. Out of the 5 dentists, one was a dental prosthetic specialist (prosthodontist), one was a gums specialist (periodontist), one was an oral surgeon, and one was an endodontist. The fifth one was a general dentist. 

Patients with a problem related to a particular specialty were referred to that specialist throught he use of this expert system. Patients whose symptoms were vague and difficult to establish a diagnosis on were referred to the general dentist for evaluation. This system helped manage patient traffic without them forming long lines to first get a referral and then get into another line to be treated by the specialist. 

```python
# Dental Camp Expert System with Likely Diagnosis

def dental_camp_expert_system():
    print("Welcome to the Dental Camp Expert System!")
    
    symptoms = input("List out your patients symptoms in this format: 'bleeding from gums', 'pain', 'fever', 'swelling', 'cavity', etc ").lower()
    
    # Symptom categories and counters
    gum_bleeding = 'bleeding from gums'
    pain = 'pain'
    fever = 'fever'
    swelling = 'swelling'
    cavity = 'hole' or 'cavity'
    symptoms_list = []
    reversible_pulpitis_symptoms = []
    rps_counter = 0
    irreversible_pulpitis_symptoms = []
    irps_counter = 0
    pericoronitis_symptoms = []
    pc_counter = 0
    pulpal_necrosis_symptoms = []
    pn_counter = 0
    caries_symptoms = []
    caries_counter = 0

    likely_diagnosis = []

    # Logic for referring to periodontist
    if gum_bleeding in symptoms:
        likely_diagnosis.append("Gingivitis or Periodontitis")
        print("Likely Diagnosis: Gingivitis or Periodontitis")
        print("Go see the periodontist.")
        return
    
    # Pain-related logic
    if pain in symptoms:
        symptoms_list.append("Pain")
        pain_site = input("Where does the pain occur? ").lower()
        pain_onset = input("When does the pain start? ").lower()
        pain_character = input("Is it dull ('dubao') or sharp ('Chubnay wali')? ").lower()
        pain_radiation = input("Does it stay localized or does it spread to your head and/or neck? ").lower()
        pain_aggravators = input("Is it after eating something hot/cold/sugary or spontaneous? ").lower()
        pain_duration = input("How long does it last? ").lower()
        pain_severity = int(input("How severe is the pain on a scale of 1 to 10? "))
        pain_relievers = input("Does it improve after eating painkillers? ").lower()
        
        if "one tooth" in pain_site:
            reversible_pulpitis_symptoms.append("Localized pain")
            rps_counter += 1

        if "third molar area" in pain_site or "back of mouth" in pain_site:
            pericoronitis_symptoms.append("Pain in third molar region")
            pc_counter += 1

        if "unable to tell which tooth" in pain_site or "spreads to head or neck" in pain_radiation:
            irreversible_pulpitis_symptoms.append("Non-localized pain")
            irps_counter += 1
        if "spreads to head or neck" in pain_radiation:
            pericoronitis_symptoms.append("Radiation")
            pc_counter += 1

        if "after chewing" in pain_onset:
            reversible_pulpitis_symptoms.append("Pain on biting")
            rps_counter += 1
            irreversible_pulpitis_symptoms.append("Pain on biting")
            irps_counter += 1

        if "spontaneous" in pain_onset:
            irreversible_pulpitis_symptoms.append("Spontaneous pain")
            irps_counter += 1
            pericoronitis_symptoms.append("Spontaneous Pain")
            pc_counter += 1

        if "cold food" in pain_aggravators or "hot food" in pain_aggravators:
            reversible_pulpitis_symptoms.append("Pain on cold/hot stimulus")
            rps_counter += 1
            caries_symptoms.append("Pain on hot/cold stimulus")
            caries_counter += 1
            irreversible_pulpitis_symptoms.append("Pain on cold/hot stimulus")
            irps_counter += 1

        if "dull" in pain_character:
            pulpal_necrosis_symptoms.append("Dull ache")
            pn_counter += 1
            
            pericoronitis_symptoms.append("Dull ache")
            pc_counter += 1

        if "sharp" in pain_character:
            irreversible_pulpitis_symptoms.append("Sharp pain")
            irps_counter += 1

        if "lasts for a few seconds" in pain_duration:
            reversible_pulpitis_symptoms.append("Fleeting pain")
            rps_counter += 1

        if "lasts for a while" in pain_duration:
            irreversible_pulpitis_symptoms.append("Lasting pain")
            irps_counter += 1
            
            pericoronitis_symptoms.append("Lasting pain")
            pc_counter += 1

        if pain_severity > 5:
            irreversible_pulpitis_symptoms.append("Severe pain")
            irps_counter += 1
            
            pericoronitis_symptoms.append("Severe Pain")
            pc_counter = +1

        if pain_severity <= 5:
            reversible_pulpitis_symptoms.append("Not so severe")
            rps_counter += 1
            

        if "yes" in pain_relievers:
            reversible_pulpitis_symptoms.append("Relieved by NSAIDs")
            rps_counter += 1
            pericoronitis_symptoms.append("Relieved by NSAIDs")
            pc_counter += 1

        if "no" in pain_relievers:
            irreversible_pulpitis_symptoms.append("Not relieved by NSAIDs")
            irps_counter += 1

    # Combined symptoms logic
    if pain and fever and swelling in symptoms:
        pulpal_necrosis_symptoms.append("Pain, fever, swelling")
        pn_counter += 1
        
    if pain and fever in symptoms:
        symptoms_list.append("Pain and fever")

    # Logic for cavities
    if cavity in symptoms:
        caries_symptoms.append("Cavitated tooth")
        caries_counter += 1

    # Decision logic based on counters
    # if pn_counter > 0:
    #     likely_diagnosis.append("Pulpal Necrosis or Abscess")
    #     print(pulpal_necrosis_symptoms)
    #     print("Likely Diagnosis: Pulpal Necrosis or Abscess")
    #     print("Go see the General Dentist")
    elif pc_counter > 0:
        likely_diagnosis.append("Pericoronitis")
        print(pericoronitis_symptoms)
        print("Likely Diagnosis: Pericoronitis")
        print("Go see the Oral Surgeon.")
    elif irps_counter > rps_counter and irps_counter > 0:
        likely_diagnosis.append("Irreversible Pulpitis")
        print(irreversible_pulpitis_symptoms)
        print("Likely Diagnosis: Irreversible Pulpitis")
        print("Go see the endodontist.")
    elif rps_counter > 0:
        likely_diagnosis.append("Reversible Pulpitis")
        print(reversible_pulpitis_symptoms)
        print("Likely Diagnosis: Reversible Pulpitis")
        print("Go see the endodontist.")
    elif caries_counter > 0:
        likely_diagnosis.append("Dental Caries")
        print(caries_symptoms)
        print("Likely Diagnosis: Dental Caries")
        print("Go see the general dentist.")
    else:
        likely_diagnosis.append("Unclear symptoms")
        print(symptoms)
        print("Likely Diagnosis: Unclear symptoms")
        print("Your symptoms are unclear. Please see the general dentist for further evaluation.")

# Run the expert system
dental_camp_expert_system()
```
